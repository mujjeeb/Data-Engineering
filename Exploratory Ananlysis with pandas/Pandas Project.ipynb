1. How many employees are there in the dataset?

```python
nr_employees = str(df.shape[0])
print("There are " +nr_employees+ " employees in the dataset.")
```

2. what is the highest salary among the employees?

```python
max_salary_earner = df.loc[df['Salary']==(max(df.Salary))]
print(f"The highest salary among the employees is {max_salary_earner['Salary'].to_string(index=False)} earned by {max_salary_earner['First Name'].to_string(index=False)} {max_salary_earner['Surname'].to_string(index=False)} ")
```

3. Which department has the highest average salary?

```python
df_department = df.groupby('Department')['Salary'].mean().reset_index()
dept_with_highest_avg = df_department[df_department['Salary'] == max(df_department['Salary'])]['Department'].to_string(index=False)
highest_avg_salary = max(df_department['Salary'])
print(f"The department with the highest average salary is {dept_with_highest_avg} with {highest_avg_salary}")
```

4. What is the total salary expenditure for the company?

```python
total_salary = df['Salary'].sum()
print(f"The total salary expenditure for the company is {total_salary}")
```

5. How many employees are in each department?

```python
df_department_sum = df.groupby('Department')['Department'].count()
print(df_department_sum)
```
